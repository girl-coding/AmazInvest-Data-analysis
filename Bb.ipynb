```python
weight_rating = 0.4
weight_price = 0.3
weight_num_rating = 0.4


# Multiply each rank by its respective weight
df['weighted_rating_rank'] = df['Rank_Rating'] * weight_rating
df['weighted_price_rank'] = df['Rank_Price'] * weight_price
df['weighted_num_rating_rank'] = df['Rank_Number of Ratings'] * weight_num_rating

# Calculate the overall rank by summing the weighted ranks and dividing by the number of criteria
df['Rank'] = (df['weighted_rating_rank'] + df['weighted_price_rank'] + df['weighted_num_rating_rank']) / 3

# Sort the DataFrame by the overall rank in ascending order
#df = df.sort_values('Rank')

# Reset the index of the DataFrame
df = df.reset_index(drop=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

weights = {
    'Rating': 0.3,
    'Number of Ratings': 0.2,
    'Price': 0.1,
    'Minimum Price': 0.05,
    'Maximum Price': 0.05,
    'Average Price': 0.05,
    'Total of Sellers': 0.05,
    'Rank': 0.1
}

# Preprocess 'Number of Ratings' column
dup['Number of Ratings'] = dup['Number of Ratings'].str.replace(',', '').astype(float)

# Normalize the numerical features in dup
numerical_features = ['Rating', 'Number of Ratings', 'Price', 'Minimum Price',
                      'Maximum Price', 'Average Price', 'Total of Sellers', 'Rank']
scaler = MinMaxScaler()
dup[numerical_features] = scaler.fit_transform(dup[numerical_features])

# Calculate the weighted sum
weighted_sum = sum(dup[feature] * weight for feature, weight in weights.items())

# Scale the score to a range of 0-10
max_weighted_sum = sum(weights.values())
score_of_interest = (weighted_sum / max_weighted_sum) * 10

# Assign the calculated score_of_interest to df without introducing NaN values
df['Score of Interest'] = np.nan
dup['Score of Interest'] = score_of_interest
df.loc[dup.index, 'Score of Interest'] = score_of_interest

# Perform additional processing as needed
min_score = df['Score of Interest'].min()
max_score = df['Score of Interest'].max()

df['Score of Interest'] = ((df['Score of Interest'] - min_score) / (max_score - min_score)) * 10
```

```python
scaler = MinMaxScaler(feature_range=(3, 12))
df['Score of Interest'] = scaler.fit_transform(df['Score of Interest'].values.reshape(-1, 1))
```

```python
import pandas as pd

df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Minimum Price'] = pd.to_numeric(df['Minimum Price'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Number of Ratings'] = pd.to_numeric(df['Number of Ratings'], errors='coerce')

df['Demand Potential'] = 0
df['Competition Score'] = 0

# # # # Calculate demand potential 
df['Demand Potential'] = ((df['Rating'] * df['Rank']) / (df['Rating'].max() * df['Rank'].max())) * 20

# # # # Calculate competition score
df['Competition Score'] = ((df['Rank'].max() - df['Rank']) / df['Rank'].max()) * 10

# # # # Perform min-max normalization for niche score 
min_score = min(df['Demand Potential'].min(), df['Competition Score'].min()) 
max_score = max(df['Demand Potential'].max(), df['Competition Score'].max())

df['Niche Score'] = ((df['Demand Potential'] - min_score) / (max_score - min_score)) * 20

from sklearn.preprocessing import MinMaxScaler

# Calculate min and max values
min_score = df['Niche Score'].min()
max_score = df['Niche Score'].max()

# Scale the 'Niche Score' column
scaler = MinMaxScaler(feature_range=(3, 12))
df['Niche Score'] = scaler.fit_transform(df['Niche Score'].values.reshape(-1, 1))


```

```python

# Generate investment description
for index, row in df.iterrows():
    demand = row['Demand Potential']
    competition = row['Competition Score']
    niche_score = row['Niche Score']

    description = ""

    # Determine the state of demand
    if demand < 3:
        description += "Low demand. "
    elif demand < 7:
        description += "Moderate demand. "
    else:
        description += "High demand. "

    # Determine the state of competition
    if competition < 3:
        description += "Low competition. "
    elif competition < 7:
        description += "Moderate competition. "
    else:
        description += "High competition. "

    # Provide insights about investment potential and risks
    if niche_score < 3:
        description += "Investment may not be favorable due to low demand and high competition. Consider the associated risks."
    elif niche_score < 7:
        description += "Investment potential is moderate considering the balance between demand and competition. Assess the risks involved."
    else:
        description += "Investment opportunity seems promising with high demand and low competition. Evaluate potential risks."

    df.at[index, 'Investment Description'] = description
```
